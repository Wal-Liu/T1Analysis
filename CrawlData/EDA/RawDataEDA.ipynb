# TimeZone convert

```python
matchHistory_DateTime = pd.read_csv('../Output/matchHistory_DateTime.csv', sep = ',')
matchHistory_DateTime
```

```python
matchHistory['date_time'] = matchHistory_DateTime['date']
```

```python
matchHistory['date_time'] = pd.to_datetime(matchHistory['date_time'])
matchHistory['date_time'] = matchHistory['date_time'].dt.tz_localize('UTC')
matchHistory['date_time_VietNam'] = matchHistory['date_time'].dt.tz_convert('Asia/Ho_Chi_Minh')
matchHistory['date_time_Korea'] = matchHistory['date_time'].dt.tz_convert('Asia/Seoul')

matchHistory
```

```python
matchHistory.to_csv('../Output/T1_Match_History.csv', index = False)
```
